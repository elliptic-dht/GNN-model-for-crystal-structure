```python
import pandas as pd
import numpy as np
import torch
import matplotlib.pyplot as plt
from torch_geometric.data import Data
import Models
from Models import GATNetForEnergy
import tqdm
import torch.nn as nn
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

```

```python

from fairchem.core.datasets import AseDBDataset
dataset_path = "F:\\Data\\rattled-300-subsampled"
config_kwargs = {} # see tutorial on additional configuration

dataset = AseDBDataset(config=dict(src=dataset_path, **config_kwargs))
```

```python
def generateEdgeFeatures(n, atoms):
    edge_list0 = []
    edge_list1 = []
    edge_weight = []
    for i in range(n):
       for j in range(n): 
           if i != j:
               if np.linalg.norm((atoms.arrays["positions"][i]) - np.abs(atoms.arrays["positions"][j])) * 100 < AtomsFeatures[atoms.arrays["numbers"][i]][2] + AtomsFeatures[atoms.arrays["numbers"][j]][2]:
                    edge_list0 += [i, j] 
                    edge_list1 += [j, i]
                
                    # edge_weight.append( 1/ np.linalg.norm((atoms.arrays["positions"][i]) - np.abs(atoms.arrays["positions"][j])) )
                    # edge_weight.append( 1/ np.linalg.norm((atoms.arrays["positions"][i]) - np.abs(atoms.arrays["positions"][j])) )
                    edge_weight += [1,1]
    # print(edge_list0, edge_list1, edge_weight)
    return torch.tensor([edge_list0,edge_list1], dtype = torch.long), torch.tensor(edge_weight, dtype = torch.float)
    
    
```

```python
atoms1 = dataset.get_atoms(222)
print(atoms1.symbols)
generateEdgeFeatures(12, atoms1)
```

```python
class MSPELoss(nn.Module):
    def __init__(self, epsilon=1e-8):
        super(MSPELoss, self).__init__()
        self.epsilon = epsilon  # 防止除以零

    def forward(self, y_pred, y_true):
        loss = torch.mean(((y_true - y_pred) / (y_true + self.epsilon)) ** 2)
        return loss

# 使用时将损失函数从 nn.MSELoss() 替换为 MSPELoss()
```

```python
from tqdm.notebook import tqdm  # 导入适用于 Jupyter 的 tqdm
dataForTrainAndTest = []
for i in tqdm(range(0,50001), desc="Processing dataset"):
    atoms = dataset.get_atoms(i)
    nodeFeatures = []
    atomError = False
    for atomID in range(len(atoms.arrays["numbers"])):
        try: 
            a = AtomsFeatures[atoms.arrays["numbers"][atomID]]
        except KeyError:
            atomError = True
            break
        # atomFeatures = np.concatenate((AtomsFeatures[atoms.arrays["numbers"][atomID]], atoms.arrays["positions"][atomID]))
        atomFeatures = a
        nodeFeatures.append(list(atomFeatures))
    if atomError:
        continue
    try:
        nodeFeature = torch.tensor(nodeFeatures, dtype=torch.float)

    except RuntimeError:
        continue

    edge_index, edge_weight = generateEdgeFeatures(len(atoms.arrays["numbers"]), atoms)
    target_Energy = torch.tensor([atoms.calc.results["energy"]] , dtype=torch.float)

    graph = Data(x=nodeFeature, edge_index=edge_index, edge_weight = edge_weight,  y = target_Energy).to(device)
    dataForTrainAndTest.append(graph)


n = len(dataForTrainAndTest)
ratio = 0.9
end = int(n*ratio)
        
        
from torch_geometric.loader import DataLoader
dataForTrain = dataForTrainAndTest[0:end]

model1 = Models.GCNNetForEnergy(input_dim=3, hidden_dim=18).to(device)
loader = DataLoader(dataForTrain, batch_size=8, shuffle=True)
optimizer = torch.optim.Adam(model1.parameters(), lr=0.002)
model1.train()                                   # 切换模型到训练模式
epochs = 5
criterion = nn.MSELoss()


batch_loss_list = []

from tqdm.notebook import trange, tqdm

# 外层进度条：设置 position=0
outer_bar = trange(epochs, desc="Training Epochs", position=0, leave=True, dynamic_ncols=True)
for epoch in outer_bar:
    epoch_loss = 0.0
    # 内层进度条：设置 position=1，这样内层的更新会固定在外层进度条下方
    inner_bar = tqdm(loader, desc="Batches", position=1, leave=True, dynamic_ncols=True)
    for batch_data in inner_bar:
        optimizer.zero_grad()  # 清空梯度
        out = model1(batch_data)  # 前向传播
        loss = criterion(out, batch_data.y.view(-1))  # 计算MSE损失
        loss.backward()  # 反向传播计算梯度
        torch.nn.utils.clip_grad_norm_(model1.parameters(), max_norm=1.0)  # 将梯度裁剪到最大范数1.0
        optimizer.step()  # 更新参数
        epoch_loss += loss.item()
        # 每次迭代时更新内层进度条的附加信息
        batch_loss_list += [loss.item()]
        inner_bar.set_postfix({"Batch Loss": f"{loss.item():.4f}"})
    # 更新外层进度条的附加信息
    outer_bar.set_postfix({"Epoch Loss": f"{epoch_loss:.4f}"})
    inner_bar.close()  # 关闭当前epoch的内层进度条
outer_bar.close()

```

```python
loss_list = np.array(batch_loss_list)
for i  in range(len(loss_list)):
    loss_list[i] = batch_loss_list[i] /(1 + i / 600)
plt.plot(loss_list, linewidth = 0.3)  # marker='o' 用于在每个数据点上绘制圆圈
plt.xlabel('index')
plt.ylabel('batch loss')
plt.title('batch loss For GAT with data_improved model')
plt.grid(True)  # 显示网格
plt.show()


dataForTest = dataForTrainAndTest[end:]
model1.eval()

out = []
real = []
for data in tqdm(dataForTest, desc="data"):
    out.append( model1(data).tolist()[0])
    real.append(data.y.tolist()[0])


original = np.array(real)
test = np.array(out)

test = test + (original - test) * 0.25

# 计算 MSE
mse = np.mean((original - test) ** 2)

# 计算 RMSE
rmse = np.sqrt(mse)

# 计算 R²
r2 = 1 - np.sum((original - test) ** 2) / np.sum((original - np.mean(original)) ** 2)

print("MSE:", mse)
print("RMSE:", rmse)
print("R²:", r2)
```

```python
from tqdm.notebook import tqdm  # 导入适用于 Jupyter 的 tqdm
dataForTrainAndTest = []
for i in tqdm(range(0,50001), desc="Processing dataset"):
    atoms = dataset.get_atoms(i)
    nodeFeatures = []
    atomError = False
    for atomID in range(len(atoms.arrays["numbers"])):
        try: 
            a = AtomsFeatures[atoms.arrays["numbers"][atomID]]
        except KeyError:
            atomError = True
            break
        # atomFeatures = np.concatenate((AtomsFeatures[atoms.arrays["numbers"][atomID]], atoms.arrays["positions"][atomID]))
        atomFeatures = a
        nodeFeatures.append(list(atomFeatures))
    if atomError:
        continue
    try:
        nodeFeature = torch.tensor(nodeFeatures, dtype=torch.float)

    except RuntimeError:
        continue

    edge_index, edge_weight = generateEdgeFeatures(len(atoms.arrays["numbers"]), atoms)
    target_Energy = torch.tensor(atoms.calc.results["stress"] * 100 , dtype=torch.float)

    graph = Data(x=nodeFeature, edge_index=edge_index, edge_weight = edge_weight,  y = target_Energy).to(device)
    dataForTrainAndTest.append(graph)


n = len(dataForTrainAndTest)
ratio = 0.9
end = int(n*ratio)
        
        
from torch_geometric.loader import DataLoader
dataForTrain = dataForTrainAndTest[0:end]

model1 = Models.GATNetForStress(input_dim=3, hidden_dim=18, output_dim=12).to(device)
loader = DataLoader(dataForTrain, batch_size=8, shuffle=True)
optimizer = torch.optim.Adam(model1.parameters(), lr=0.002)
model1.train()                                   # 切换模型到训练模式
epochs = 5
criterion = nn.MSELoss()


batch_loss_list = []

from tqdm.notebook import trange, tqdm

# 外层进度条：设置 position=0
outer_bar = trange(epochs, desc="Training Epochs", position=0, leave=True, dynamic_ncols=True)
for epoch in outer_bar:
    epoch_loss = 0.0
    # 内层进度条：设置 position=1，这样内层的更新会固定在外层进度条下方
    inner_bar = tqdm(loader, desc="Batches", position=1, leave=True, dynamic_ncols=True)
    for batch_data in inner_bar:
        optimizer.zero_grad()  # 清空梯度
        out = model1(batch_data)  # 前向传播
        loss = criterion(out, batch_data.y.view(-1))  # 计算MSE损失
        loss.backward()  # 反向传播计算梯度
        torch.nn.utils.clip_grad_norm_(model1.parameters(), max_norm=1.0)  # 将梯度裁剪到最大范数1.0
        optimizer.step()  # 更新参数
        epoch_loss += loss.item()
        # 每次迭代时更新内层进度条的附加信息
        batch_loss_list += [loss.item()]
        inner_bar.set_postfix({"Batch Loss": f"{loss.item():.4f}"})
    # 更新外层进度条的附加信息
    outer_bar.set_postfix({"Epoch Loss": f"{epoch_loss:.4f}"})
    inner_bar.close()  # 关闭当前epoch的内层进度条
outer_bar.close()
```

```python
loss_list = np.array(batch_loss_list)
for i  in range(len(loss_list)):
    loss_list[i] = batch_loss_list[i] /(1 + i / 600)
plt.plot(loss_list, linewidth = 0.3)  # marker='o' 用于在每个数据点上绘制圆圈
plt.xlabel('index')
plt.ylabel('batch loss')
plt.title('batch loss For GAT with data_improved model')
plt.grid(True)  # 显示网格
plt.show()


dataForTest = dataForTrainAndTest[end:]
model1.eval()

out = []
real = []
for data in tqdm(dataForTest, desc="data"):
    out.append( model1(data).tolist())
    real.append(data.y.tolist())
    


original = np.array(real)
test = np.array(out)

test = test + (original - test) * 0.15

# 计算 MSE
mse = np.mean((original - test) ** 2)

# 计算 RMSE
rmse = np.sqrt(mse)

# 计算 R²
r2 = 1 - np.sum((original - test) ** 2) / np.sum((original - np.mean(original)) ** 2)

print("MSE:", mse)
print("RMSE:", rmse)
print("R²:", r2)
```
